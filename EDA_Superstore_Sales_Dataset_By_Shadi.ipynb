## **UNIVARIATE ANALYSIS**

```python
# Set the style
sns.set(style="whitegrid")

# Numerical columns
numerical_cols=['Row ID','Postal Code','Sales']

# Loop through numerical columns
for col in numerical_cols:
    plt.figure(figsize=(14, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Categorical columns
categorical_cols=[
'Ship Mode',
'Segment',
'Country',
'Region',
'Category',
'Sub-Category'
]

# Plot countplots
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## **Bivariate Analysis**

```python
# Categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Country', 'Region', 'Category', 'Sub-Category']

# Countplots by Sales
for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue='Sales', palette='husl')
    plt.title(f'{col} by Sales')
    plt.tight_layout()
    plt.show()
```

```python
# Loop through combinations of numerical and categorical columns
for cat_col in categorical_cols:
        plt.figure(figsize=(6, 4))
        sns.boxplot(x=cat_col, y='Sales', data=df, palette='coolwarm')
        plt.title(f'Sales by {cat_col}')
        plt.tight_layout()
        plt.show()
```

## **Multivariate analysis**

```python
# Correlation heatmap
numerical_cols = ['Row ID', 'Postal Code', 'Sales']
plt.figure(figsize=(10, 6))
# Calculate the correlation matrix for the numerical columns
correlation_matrix = df[numerical_cols].corr()
# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
