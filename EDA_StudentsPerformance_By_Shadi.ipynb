# **EDA**

## **Univariate Analysis**

```python
# Set the style
sns.set(style="whitegrid")

# Numerical columns
numerical_cols = ['math score', 'reading score', 'writing score']

# Loop through numerical columns
for col in numerical_cols:
    plt.figure(figsize=(14, 5))

    # Histogram
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')

    # Boxplot
    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col], color='lightcoral')
    plt.title(f'Boxplot of {col}')

    plt.tight_layout()
    plt.show()
```

```python
# Categorical columns
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Plot countplots
for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=df, x=col, order=df[col].value_counts().index, palette='Set2')
    plt.title(f'Countplot of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

## **Bivariate Analysis**

```python
# Categorical columns
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Countplots by math score
print('Countplots by math score\n')
for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue='math score', palette='husl')
    plt.title(f'{col} by math score')
    plt.tight_layout()
    plt.show()

print('Countplots by reading score\n')
# Countplots by math score
for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue='reading score', palette='husl')
    plt.title(f'{col} by reading score')
    plt.tight_layout()
    plt.show()

print('Countplots by writing score\n')
# Countplots by writing score
for col in categorical_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=col, hue='writing score', palette='husl')
    plt.title(f'{col} by writing score')
    plt.tight_layout()
    plt.show()
```

```python
# Loop through combinations of numerical and categorical columns
for cat_col in categorical_cols:
    for num_col in numerical_cols:
        plt.figure(figsize=(10, 4))
        sns.boxplot(x=cat_col, y=num_col, data=df, palette='coolwarm')
        plt.title(f'{num_col} by {cat_col}')
        plt.tight_layout()
        plt.show()
```

## **Multivariate analysis**

```python
# Pairplot to see multivariate relationship among numerical features and the target
sns.pairplot(df[numerical_cols], hue='math score', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features Colored by math score', y=1.02)
plt.show()

print('Pairplot to see multivariate relationship among numerical features and the target\n')

sns.pairplot(df[numerical_cols], hue='reading score', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features Colored by reading score', y=1.02)
plt.show()

print('Pairplot to see multivariate relationship among numerical features and the target\n')

sns.pairplot(df[numerical_cols], hue='writing score', palette='husl', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Features Colored by writing score', y=1.02)
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
# Calculate the correlation matrix for the numerical columns
correlation_matrix = df[numerical_cols].corr()
# Plot the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()

```
