## **4. Data Modification and Creation**

```python
# 31. Add new column 'Bonus' 10% of Salary
def calculateBonus(salary):
  return salary * 0.1

print('Add new column Bonus 10% of Salary')
df['Bonus']=df['Salary'].apply(calculateBonus)
print(df)
print('-----------------------------------------------\n')

# 32. Add column 'Net Salary'= Salary + Bonus
print('Add column Net Salary= Salary + Bonus')
df['Net Salary']=df['Salary']+df['Bonus']
print(df)
print('-----------------------------------------------\n')

# 33. Create 'Age Group' : 'Young' if Age<30, Old otherwise
def ageGroup(age):
  if age<30:
    return 'Young'
  else:
    return 'Old'

print('Create Age Group : Young if Age<30, Old otherwise')
df['Age Group']=df['Age'].apply(ageGroup)
print(df)
print('-----------------------------------------------\n')

# 34.Convert all columns names to lowercase
print('Convert all columns names to lowercase')
df.columns=df.columns.str.lower()
print(df)
print('-----------------------------------------------\n')

#35. Rename the column 'Salary' to Base Salary'
print('Rename the column Salary to Base Salary')
df.rename(columns={'salary':'Base Salary'},inplace=True)
print(df)
print('-----------------------------------------------\n')

#36. Delete the column 'Net Salary'
print('Delete the column Net Salary')
df=df.drop('net salary',axis=1)
print(df)
print('-----------------------------------------------\n')

#37. Delete the last row
print('Delete the last row')
df=df.drop(df.index[-1])
print(df)
print('-----------------------------------------------\n')

#38. Reindexing columns so 'Name' comes last
print('Reindexing columns so Name comes last')
df = df.reindex(columns=['age','gender','department','Base Salary','bonus','experience','age group','name'])
print(df)
print('-----------------------------------------------\n')

# 39. Add a new row for new employee
print('Add a new row for new employee')
df.loc[len(df)] = [36, 'Male', 'IT', 7500, 14, 14, 'Young', 'Shadi']
print(df)
print('-----------------------------------------------\n')

# 40. Add a column showing the length of each name
print('Add a column showing the length of each name')
df['Name Length'] = df['name'].apply(len)
print(df)
```

## **7. Advanced Operations**

```python
#56. Use '.Apply()' to compute sequared age in a new coulmn
print('Use .Apply() to compute sequared age in a new coulmn\n')
df['Squared Age']=df['age'].apply(lambda x:x**2)
print(df)
print('\n-----------------------------------------------\n')

#57. Use '.replace()' to convert 'Gender' to numeric codes (Male: 0,Female :1)
print('Use .replace() to convert Gender to numeric codes (Male: 0,Female :1)\n')
df['gender']=df['gender'].replace({'Male':0,'Female':1})
print(df)
print('\n-----------------------------------------------\n')

#58. Create column 'Tax' = 5% of Salary if Salary>5000
print('Create column Tax = 5% of Salary if Salary>5000\n')
df['Tax']=df['Base Salary'].apply(lambda x:x*0.05 if x>5000 else 0)
print(df)
print('\n-----------------------------------------------\n')

#59. Calculate the range (Max-Min) of salaries
print('Calculate the range (Max-Min) of salaries\n')
print(df['Base Salary'].max()-df['Base Salary'].min())
print('\n-----------------------------------------------\n')

#60. Compute the mean salary after removing outliers (using IQR)
print('Compute the mean salary after removing outliers (using IQR)\n')
print('Mean salary Befor removing outliers: ',df['Base Salary'].mean())

# Step 1: Calculate Q1 and Q3
Q1 = df['Base Salary'].quantile(0.25)
Q3 = df['Base Salary'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds to filter out outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print('lower bound: ',lower_bound)
print('upper bound: ',upper_bound)

# Step 3: Filter out the outliers
filtered_df = df[(df['Base Salary'] >= lower_bound) & (df['Base Salary'] <= upper_bound)]

# Step 4: Compute mean salary without outliers
mean_salary = filtered_df['Base Salary'].mean()

print("Mean salary after removing outliers:", mean_salary)
print('\n-----------------------------------------------\n')
```
